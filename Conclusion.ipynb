## Question 1:


K-means is a popular unsupervised learning algorithm that is widely used for clustering data into k groups or clusters. The algorithm is based on the idea of minimizing the sum of the distances between the data points and the centroids of the clusters they belong to.

The World Health Organization (WHO) has a dataset called 'SEARO' which stands for South-East Asia Region, that contains a wide range of health-related information for the countries of the Southeast Asia region. 

When using k-means to analyze this dataset, it is important to note that the scale of the different features can have a significant impact on the results of the algorithm. For example, if the scale of the different features is not taken into account, the algorithm might give more importance to the features with higher values, even if they are not the most relevant for the analysis.

To address this issue, the data can be scaled before applying k-means. Scaling the data ensures that all the features have the same weight in the analysis, and it can also improve the performance of the algorithm. There are several ways to scale the data, such as Min-Max scaling, standardization, and normalization.

When I use k-means on the WHO's SEARO dataset, I used Min-Max scaling, this method scales the data to a fixed range between 0 and 1. The results of the k-means algorithm on the scaled data showed slightly different clusters than the unscaled data. The cluster that was created by the scaled data was more compact and had a more homogeneous distribution of data points.

In conclusion, k-means is a useful tool for analyzing the WHO's SEARO dataset as it allows identifying patterns and trends in the data. However, it is important to take the scale of the different features into account before applying the algorithm. Scaling the data can have a significant impact on the results of the k-means algorithm and can improve the performance of the algorithm.

## Question 2:

Hierarchical Birch is a clustering algorithm that is similar to k-means but with some key differences that make it useful in certain scenarios.

One of the main advantages of Hierarchical Birch over k-means is its ability to handle large datasets with a high number of dimensions. k-means has a high computational cost for high-dimensional data, as it requires calculating distances between all data points and the centroids of the clusters. Hierarchical Birch, on the other hand, uses a hierarchical structure to reduce the computational cost and can handle large datasets more efficiently.

Another advantage of Hierarchical Birch is that it can handle non-spherical clusters. k-means assumes that clusters are spherical, which can lead to poor results when the clusters have a different shape. Hierarchical Birch can handle clusters of different shapes and can also identify subclusters within clusters.

Furthermore, Hierarchical Birch can handle noise and outliers in the data more effectively than k-means. k-means is sensitive to noise and outliers and they can have a significant impact on the results of the algorithm. Hierarchical Birch uses a CF tree to store the data points and it can handle noise and outliers by creating subclusters within clusters.

In summary, Hierarchical Birch can be useful for clustering large datasets with high dimensional data, non-spherical clusters, and noise or outliers. It is a more robust algorithm than k-means and can handle these scenarios more effectively.
